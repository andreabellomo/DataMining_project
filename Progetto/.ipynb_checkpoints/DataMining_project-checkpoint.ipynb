```python
import pandas as pd
import matplotlib.pyplot as plot
from math import sqrt
```

```python
brca_mutation = pd.read_csv('brca_mutations.txt', delimiter = "\t")
brca_mutation= brca_mutation.transpose()

brca_mutation.index=[x[:12] for x in brca_mutation.index]
#brca_mutation=brca_mutation.reindex([x[:12] for x in brca_mutation.index])
brca_mutation.head()
```

# Pre-processing sui dataframe
Viene effettuato un pre-proccesing su i due dataframe eliminando i record che non sono in comune e risolvendo la presenza di alcuni vaori NaN.

```python

idx1 = pd.Index([x[:12] for x in brca_mutation.index])
idx2 = pd.Index(brca_clinical.index)

list_mutation=set(list(brca_mutation.index))
list_clinical=set(list(brca_clinical.index))
list_intersection = list_mutation & list_clinical

del_mutation=set(list_mutation)-set(list_intersection)
brca_mutation.drop(del_mutation,axis=0,inplace=True)


del_clinical=set(list_clinical)-set(list_intersection)
brca_clinical.drop(del_clinical,axis=0,inplace=True)
```

# Riduzione della dimensionalitÃ  con PCA

La PCA Ã¨ influenzata dalla differenza di scala nei dati, per cui conviene normalizzare le feature. Utilizziamo lo StandardScaler di Scikit-Learn.



```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_norm = ss.fit_transform(brca_mutation)
mean, var = ss.mean_, ss.var_
X_norm
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=15)

pcomps = pca.fit_transform(X_norm)
pcomps
```

```python
final_df = pd.DataFrame(data=pcomps, columns=['PC%s' % _ for _ in range(15)])
final_df.index=[x[:12] for x in brca_mutation.index]
```

# Performance

Per la misura delle performance utilizzeremo la <B>Jaccard Index (J)</B> e <B> la Fowlkes Mallows Index (FM)</B>.
L'indice <B> Fowlkesâ€“Mallows </B> Ã¨ un metodo di valutazione esterno utilizzato per determinare la somiglianza tra due raggruppamenti (cluster ottenuti). Un valore piÃ¹ alto per l'indice Fowlkesâ€“Mallows indica una maggiore somiglianza tra i cluster  o le classificazioni.


```python
import seaborn as sns
from sklearn.manifold import TSNE

def grafico(df_predetto):
    
    X = df_predettoK.drop('cluster', axis=1).to_numpy().astype(np.float64)
    
    X_emb = TSNE(n_components=2, learning_rate=.001, init='random').fit_transform(X)
    
    df = pd.DataFrame(data=X_emb, columns=['PC1', 'PC2'])
    df['cluster'] = df_predetto['cluster'].to_numpy()
    sns.scatterplot(data=df, x='PC1', y='PC2',hue='cluster', palette="deep")
    
```

Definamo <b>U = {u1, u2. . .uR}</b> come l'insieme nella quale Ui Ã¨ il sottoinsieme degli elementi di classe i. <br>
Definiamo <b>V = {v1, v2. . .vC}</b> come l'insieme delle etichette predette con i relativi sottoinsiemi. <br>
Definiamo con <b>a</b> l'insieme degli elementi che sono sia in U che in V. <br>
Definiamo con <b>b</b> l'insieme degli elementi che sono nello stesso gruppo in U ma in gruppi diversi in V. <br>
Definiamo con <b>c</b> l'insieme degli elementi che sono nello stesso gruppo in U ma in gruppi diversi in V. <br>

```python
from itertools import combinations

S = set([])

brca_clinical = brca_clinical[brca_clinical['SUBTYPE'].notna()]

tumors = set(brca_clinical['SUBTYPE'])

for tumor in tumors:
    same_tumor = brca_clinical[brca_clinical['SUBTYPE'] == tumor].index
    S = S.union(set(combinations(same_tumor, 2)))
    #print(S)
    
# brca_clinical.loc['TCGA-B6-A0I5']['SUBTYPE'], brca_clinical.loc['TCGA-B6-A40B']['SUBTYPE']
```

```python

def jaccard(S, df_predetto):
    
    V = set([])

    tumors = set(df_predetto['cluster'])

    for tumor in tumors:
        same_tumor = df_predetto[df_predetto['cluster'] == tumor].index
        V = V.union(set(combinations(same_tumor, 2)))
    
    a=0
    b=0
    c=0

    for (c1,c2) in S:
        
        a+=(int)(not((brca_clinical.loc[c1]['SUBTYPE'] == brca_clinical.loc[c2]['SUBTYPE']) != (df_predetto.loc[c1]['cluster'] == df_predetto.loc[c2]['cluster'])))
    
        b+=(int)(((1) == (df_predetto.loc[c1]['cluster'] != df_predetto.loc[c2]['cluster'])))
        
    for (c1,c2) in V:
        
        c+=(int)(((1) == (brca_clinical.loc[c1]['SUBTYPE'] != brca_clinical.loc[c2]['SUBTYPE'])))

    rdab = sqrt(a+b)
    rdac = sqrt(a+c)

    return a*((rdab*rdac)/((a+b)*(a+c))) , a/(a+b+c)

    
```

# - DBSCAN

```python
print(jaccard(S, df_predettoDB))
```
