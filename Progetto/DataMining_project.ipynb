```python

idx1 = pd.Index([x[:12] for x in brca_mutation.index])
idx2 = pd.Index(brca_clinical.index)

list_mutation=set(list(brca_mutation.index))
list_clinical=set(list(brca_clinical.index))
list_intersection = list_mutation & list_clinical

del_mutation=set(list_mutation)-set(list_intersection)
brca_mutation.drop(del_mutation,axis=0,inplace=True)


del_clinical=set(list_clinical)-set(list_intersection)
brca_clinical.drop(del_clinical,axis=0,inplace=True)

print(brca_mutation)
print(brca_clinical)
```

```python
from itertools import combinations

S = set([])

brca_clinical = brca_clinical[brca_clinical['SUBTYPE'].notna()]

tumors = set(brca_clinical['SUBTYPE'])

for tumor in tumors:
    same_tumor = brca_clinical[brca_clinical['SUBTYPE'] == tumor].index
    S = S.union(set(combinations(same_tumor, 2)))
    #print(S)
    
# brca_clinical.loc['TCGA-B6-A0I5']['SUBTYPE'], brca_clinical.loc['TCGA-B6-A40B']['SUBTYPE']
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_norm = ss.fit_transform(brca_mutation)
mean, var = ss.mean_, ss.var_
X_norm
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=15)

pcomps = pca.fit_transform(X_norm)
pcomps
```

```python
final_df = pd.DataFrame(data=pcomps, columns=['PC%s' % _ for _ in range(15)])

final_df.index=[x[:12] for x in brca_mutation.index]
```

# Performance

```python
from math import sqrt
```

```python
def jaccard(S, df_predetto):
    from math import sqrt
    V = set([])

    tumors = set(df_predetto['cluster'])

    for tumor in tumors:
        same_tumor = df_predetto[df_predetto['cluster'] == tumor].index
        V = V.union(set(combinations(same_tumor, 2)))
    
    a=0
    b=0
    c=0

    for (c1,c2) in S:
        
        a+=(int)(not((brca_clinical.loc[c1]['SUBTYPE'] == brca_clinical.loc[c2]['SUBTYPE']) != (df_predetto.loc[c1]['cluster'] == df_predetto.loc[c2]['cluster'])))
    
        b+=(int)(((1) == (df_predetto.loc[c1]['cluster'] != df_predetto.loc[c2]['cluster'])))
        
    for (c1,c2) in V:
        
        c+=(int)(((1) == (brca_clinical.loc[c1]['SUBTYPE'] != brca_clinical.loc[c2]['SUBTYPE'])))

    rdab = sqrt(a+b)
    rdac = sqrt(a+c)

    return a*((rdab*rdac)/((a+b)*(a+c))) , a/(a+b+c)

    
```

```python
print(jaccard(S,df_predetto))
```
